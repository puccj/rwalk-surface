```python
import numpy as np
import matplotlib.pyplot as plt

positions = np.loadtxt("final_positions.dat")
```

```python
def to_longlat(point):
    x, y, z = point
    r = np.sqrt(x**2 + y**2 + z**2)
    lon = np.arctan2(y, x)
    lat = np.arcsin(z / r)
    return lon, lat

lon, lat = to_longlat(positions.T)
```

# Mollweide scatter plot

```python
plt.figure(figsize=(10, 5))
ax = plt.subplot(111, projection='mollweide')
ax.scatter(lon, lat, s=1)
ax.set_title("Final Positions of Random Walkers on Surface")
plt.tight_layout()
plt.show()
```
