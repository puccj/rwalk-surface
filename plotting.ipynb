```python
import numpy as np
import matplotlib.pyplot as plt

positions = np.loadtxt("final_positions.dat")
```

```python
def to_longlat(point, center=(5,5,5), radius=4.5):
    x, y, z = point
    x -= center[0]
    y -= center[1]
    z -= center[2]
    
    lon = np.arctan2(y, x)
    lat = np.arcsin(z / radius)
    
    return lon, lat

lon, lat = to_longlat(positions.T)
```

# Mollweide scatter plot

```python
plt.figure(figsize=(10, 5))
ax = plt.subplot(111, projection='mollweide')
ax.scatter(lon, lat, s=1)
ax.set_title("Final Positions of Random Walkers on Surface")
plt.tight_layout()
plt.show()
```
